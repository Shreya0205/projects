```python
import json 
import pandas as pd
```

```python

with open('neighbor-districts-modified.json') as json_data:
    graph = json.load(json_data)

    
distName = list(graph.keys())
distName.sort()
distName = {name: index+101 for index, name in enumerate(distName)}
```

```python
def generate_edges(graph):
    edges = []
 
    for node in graph:
         
        for neighbour in graph[node]:
             
            edges.append([distName[node], distName[neighbour]])
    return edges
```

```python
edges=generate_edges(graph)
```

```python
df = pd.DataFrame(edges)
df.to_csv('edge-graph.csv', index=False, header=False)
```
